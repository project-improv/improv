```python
# import packages
import numpy as np
import matplotlib.pylab as plt
```

```python
# sample arrays
a = np.arange(100)
b = np.random.random(100)
print(a)
print(b)
```

```python
# line plot
plt.figure(figsize=(8,6))
plt.plot(a,b)
```

```python
# 2d plot

c = np.vstack((a,b))

plt.figure(figsize=(5,3))

plt.imshow(c, cmap='plasma', interpolation='nearest', aspect = 'auto')
```

```python
# bar graph
plt.figure(figsize=(8,6))
plt.bar(a,b)
```

```python
# scatter plot

d = np.random.randint(0, 50, 100)
e = [np.random.random(100) * 200]

plt.figure(figsize=(8,6))
plt.scatter(a, b, c = d, s = e)
```

```python
# polar plot

plt.rcParams["figure.figsize"] = (6.4, 4.8)

r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
```

```python
# subplot

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
ax1.plot(a, b, 'tab:green')
ax1.set_title('Axis 1')
ax2.plot(a, b)
ax2.set_title('Axis 2')
ax3.plot(a, -b, 'tab:purple')
ax3.set_title('Axis 3')
ax4.plot(a, -b, 'tab:pink')
ax4.set_title('Axis 4')


for ax in fig.get_axes():
    ax.set(xlabel='x-label', ylabel='y-label')


# for ax in fig.get_axes():
#      ax.label_outer()

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.7, 
                    hspace=0.7)
```

```python
#2d plot of data from rawC

from matplotlib.pyplot import figure

rawC = np.loadtxt('raw_C.txt')
plt.imshow(rawC, cmap='Greys')

#fig = plt.gcf()
#fig.set_size_inches(20,20)
```

```python
#exploring dimensions of rawC data

sumCols = sum(rawC)
print(len(sumCols)) #should be num of columns
print(len(rawC[0])) #confirm num of cols by pulling all items from first row
 
first = [ x[0] for x in rawC] #returns list of first elements for each list/row in rawC
print(len(first)) #num of rows
```

```python
#plotting sum of neural activity over frames
plt.figure(figsize=(10, 8))
frames = np.arange(2878)
frames
plt.plot(frames, sumCols)
plt.title('Population Neural Activity Over Time')
plt.xlabel('Frames')
plt.ylabel('Sum of Activity')
```

```python
#need to exclude any value that it zero and average everything else for each column
from statistics import mean

#calculating mean for item[0] in all rows (mean activity of all active neurons in first frame)
# excluding zeros!
col = [x[0] for x in rawC]
col_no_zeros = [y for y in col if y != 0]
col_no_zeros
mean(col_no_zeros)
```

```python
#create new transposed array
flip = np.transpose(rawC) #transposing 2d array so each row/list represents a frame
len(flip) #checking length is 2878 (number of frames)

#calculating mean (excluding 0s) of each row/frame
mean_per_frame = []
for r in flip:
    mean_per_frame.append(mean(c for c in r if c!=0))
mean_per_frame

#setting fig dimensinos
plt.figure(figsize=(10, 8))

#plotting mean neural activity of ACTIVE neurons (non 0) over frames
plt.plot(frames, mean_per_frame)
plt.title('Average Neural Activity of Active Neurons Over Time')
plt.xlabel('Frames')
plt.ylabel('Activity Level')
plt.show()
```
